PRE PROCESSING FOR DATA SET -1


```python
df['asin'].unique()
```

```python
df.drop(columns = ['asin','img'],inplace = True)
```

```python
df.head()
```

```python
#identifying categories from url 
category = df['purl']
lst=[]

for i in category:
    words=i.split('/')
    lst.append(words[3])
df['category'] =lst 
df.head(5)
```

```python
#dropping url 
df.drop(columns= ['purl'],axis =1,inplace = True)
df.head(5)
```

```python
DF.duplicated().sum()
```

```python
DF.drop_duplicates()
```

there are no more outliers and duplicates in the datset so we can start our analysis

lets find out the top10 categories that have highest average market place 


```python
df1=df.groupby('category')['mrp'].mean().sort_values(ascending = False).head(10).reset_index().head(10)
df1
```

```python
#plotting a line graph 
plt.figure(figsize=(15,5))
plt.xticks(rotation=80)


sns.lineplot(x='category',y='mrp',data=df1,marker='+')
```

top10 items with maximum discount price


```python
df2=df.groupby('category')['discount'].max().sort_values(ascending =  False).head(10).reset_index()
df2
```

```python
#plotting graph
plt.figure(figsize=(12,4))
plt.xticks(rotation=90)

plt.title(' Top 10 items with maximum discount price')
sns.barplot(x='category',y='discount',data=df2, color = 'blue')
```

Top items with HighestNumber of Ratings

```python
df3=df.groupby('category')['rating','ratingTotal'].max().sort_values(by =['rating','ratingTotal'],ascending =False).head(10).reset_index()
df3
```

```python
#plotting the graph
plt.figure(figsize=(12,4))
plt.title('Top 10 items with highestnumber of rating')
sns.barplot(data=df3,x='ratingTotal',y='category',color = 'blue')
```

Trying to find uut which seller got maximum profit

```python
df['Profit']=df['mrp']-df['price']
```

```python
df.head()
```

```python
#filterng top 10 sellers with maximum profit
df4=df.groupby('seller')['Profit'].agg(['max','mean']).sort_values(by ='mean' ,ascending = False).head(10).reset_index()
df4
```

```python
#plottingthe graph with mean profit and maximum profit
plt.figure(figsize=(12,4))
plt.xticks(rotation=90)
sns.barplot(x='seller', y='max', color='blue', alpha=0.7, label='Maximum',data=df4)
sns.barplot(x='seller', y='mean', color='red', alpha=0.7, label='Mean',data=df4)
plt.xlabel('Items')
plt.ylabel('Values')
plt.legend()
```

Finding out brands having most number of products 

```python
data = []

for brand_name in DF['brand_name'].unique():
    
    c = 0
    for i in DF['brand_name']:
        if (i == brand_name):
            c += 1

    data.append([brand_name, c])
    
df5 = pd.DataFrame(data, columns = ['name','products'])

df6= df5.sort_values(by = 'products', ascending = False).head(10)
df6
```

```python
#plotting the graph
plt.figure(figsize=(12,4))
plt.xticks(rotation=100)

plt.title(' Top 10 items with maximum number of produtcs')
sns.barplot(x='name',y='products',data=df6.head(10), color = 'blue')
```

Top 10 most expensive brands


```python
df7 = DF.groupby('brand_name').mean()['marked_price'].sort_values(ascending = False).head(10)
df8 = pd.DataFrame(df7)
df8
```
